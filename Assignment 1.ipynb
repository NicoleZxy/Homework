```python
import pandas as pd
from get_hours import get_hours
```

```python
df = pd.DataFrame(columns=pd.MultiIndex.from_product([["PJM", "MISO", "CAISO"], ['onpeak', 'offpeak', '2x16H', '7x8', 'flat']]))
monthes = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
for iso, type in df.columns:
    for year in range(2018, 2024):
        df.loc[f"{year}A", (iso, type)] = get_hours(iso, type, f"{year}A")

    for year in range(2020, 2022):
        for q in range(1, 5):
            df.loc[f"{year}Q{q}", (iso, type)] = get_hours(iso, type, f"{year}Q{q}")

    for month in range(0, 12):
        df.loc[f"2022{monthes[month]}", (iso, type)] = get_hours(iso, type, f"2022{monthes[month]}")

    for i in range(1, 32):
        df.loc[f"2023-1-{i}", (iso, type)] = get_hours(iso, type, f"2023-1-{i}")
```
